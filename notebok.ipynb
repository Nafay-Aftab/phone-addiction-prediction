### Model Training

```python
df.to_csv('cleaned_df.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,precision_score,recall_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline as sk_pipeline
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as im_pipeline
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
df['Addiction_Level_Category']=pd.cut(df['Addiction_Level'],
  bins=[-np.inf, 3.5, 7.5, np.inf],
  labels=[0, 1, 2])
```

```python
df.drop(columns=['Addiction_Level'], inplace=True)
df.rename(columns={'Addiction_Level_Category': 'Addiction_Level'}, inplace=True)
```

```python
df.select_dtypes(exclude='object').corr()['Addiction_Level']
```

```python
df=df.drop(columns=['ID',
'Academic_Performance',	
'Social_Interactions',	
'Exercise_Hours'	,
'Parental_Control',	
'Screen_Time_Before_Bed',
'Family_Communication',	
'Weekend_Usage_Hours','Age','Self_Esteem','Time_on_Education','Anxiety_Level','School_Grade','Name','Location','Depression_Level','Phone_Usage_Purpose'],axis=1)
```

```python
df.head(2)
```

```python
df.select_dtypes(exclude='object').corr()['Addiction_Level']
```

```python
sns.heatmap(df.select_dtypes(exclude='object').corr(),annot=True)
plt.show()
```

```python
df=df.drop('Gender',axis=1)
```

```python
x=df.drop('Addiction_Level',axis=1)
y=df['Addiction_Level']

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2,stratify=y)
```

```python
scaler=StandardScaler()

pipe=im_pipeline(steps=[
    
    ('scaling',scaler),
    ('smote',SMOTE(random_state=42)),
    ('model', LogisticRegression(C=0.001, multi_class='multinomial',
                                    penalty=None, random_state=42,
                                    solver='newton-cg'))
])

pipe.fit(x_train, y_train)
y_pred=pipe.predict(x_test)

print('Accuracy: ', np.round(accuracy_score(y_test,y_pred)*100))
print('Precision: ', np.round(precision_score(y_test, y_pred, average='macro')*100))
print('Recall: ', np.round(recall_score(y_test, y_pred, average='macro')*100))

print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer

cv_accuracy = cross_val_score(pipe, x_train, y_train, cv=5, scoring='accuracy')
print("CV Accuracy:", cv_accuracy.mean())

cv_precision = cross_val_score(pipe, x_train, y_train, cv=5, 
                               scoring=make_scorer(precision_score, average='macro'))
print("CV Precision:", cv_precision.mean())

cv_recall = cross_val_score(pipe, x_train, y_train, cv=5, 
                            scoring=make_scorer(recall_score, average='macro'))
print("CV Recall:", cv_recall.mean())
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')

plt.title("Confusion Matrix on Test Set")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```
